```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('charge_status.csv')
```

```python
data.shape
data.head()
```

```python
listt=data['asset_id'].unique()
print(len(listt))
```

```python
df = pd.DataFrame(columns=['DeviceID', 'Cycles'])
```

```python
def value123(i):
    first=data[data['asset_id']==i][['asset_id',"charge_status"]]
    charge=np.asarray(first['charge_status'].values.tolist(), dtype=np.float32)
    lst= []
    count=0
    lst=np.diff(np.where(np.concatenate(([charge[0]], charge[:-1] != charge[1:], [True])))[0])[::2]
#     print(lst)
    for j in lst:
#     for j in data['Charge_status']:
        if j>=3:
        
        
            count+=1
    return(count)
#     print(count)
```

```python
for i in listt:
    trues = value123(i)
    df = df.append({'DeviceID':i , 'Cycles':trues}, ignore_index=True)
    print(i, trues)
```
